## 1. CrewAI: Multi-Agent Sales Workflow
Defining specialized agents for a marketing growth task.

```python
import os
from crewai import Agent, Task, Crew, Process

os.environ["OPENAI_API_KEY"] = "your_key"

# 1. Define Agents
researcher = Agent(
  role='Market Researcher',
  goal='Find the top 3 pain points for dental clinics in 2024',
  backstory='Expert in healthcare market signals and administrative bottlenecks.',
  verbose=True
)

writer = Agent(
  role='Sales Copywriter',
  goal='Draft a cold email offering Edge talent solutions to solve those pain points',
  backstory='Master of persuasive, professional outreach.',
  verbose=True
)

# 2. Define Tasks
task1 = Task(description='Research dental industry trends.', agent=researcher, expected_output="A bulleted list of 3 trends.")
task2 = Task(description='Write a 150-word email.', agent=writer, expected_output="A draft email in markdown format.")

# 3. Execute Crew
crew = Crew(
  agents=[researcher, writer],
  tasks=[task1, task2],
  process=Process.sequential
)

print("Crew ready to execute. Explain the 'Sequential' vs 'Hierarchical' process in the interview.")
```

## 2. BigQuery (GA4) SQL Practice
Simulating the complex SQL required for GA4 BigQuery exports.

```python
import pandas as pd

# Mocking a nested BigQuery response
mock_data = [
    {'event_name': 'page_view', 'event_params': [{'key': 'page_location', 'value': {'string_value': '/home'}}]},
    {'event_name': 'lead_generated', 'event_params': [{'key': 'industry', 'value': {'string_value': 'Medical'}}]},
]

print("SQL Challenge: Write a query to extract the 'industry' from 'lead_generated' events.")
sql_solution = """
SELECT 
  (SELECT value.string_value FROM UNNEST(event_params) WHERE key = 'industry') as industry
FROM `project.dataset.events_*` 
WHERE event_name = 'lead_generated'
"""
print(f"Solution Hint:\n{sql_solution}")
```

## 3. Data Privacy (PII Masking)
Practicing GDPR-compliant data handling before feeding it to an LLM.

```python
import re

def mask_pii(text):
    # Simple regex for emails and phone numbers
    text = re.sub(r'\S+@\S+', '[EMAIL]', text)
    text = re.sub(r'\d{3}-\d{3}-\d{4}', '[PHONE]', text)
    return text

sample_text = "Contact John Doe at john@dental.com or 555-123-4567."
print(f"Masked: {mask_pii(sample_text)}")
```
