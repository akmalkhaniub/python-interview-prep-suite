## 4. Automation: Model Evaluation Script

**Task:** Write a Python function that compares a new model's metrics against a production baseline.

````python
def evaluate_deployment(new_metrics, baseline_metrics):
    # Threshold: New model must be within 2% of baseline accuracy
    threshold = 0.02
    
    acc_diff = new_metrics['accuracy'] - baseline_metrics['accuracy']
    latency_diff = new_metrics['latency_ms'] - baseline_metrics['latency_ms']
    
    passed = True
    messages = []
    
    if acc_diff < -threshold:
        passed = False
        messages.append(f"Accuracy dropped significantly: {acc_diff:.4f}")
        
    if latency_diff > 50: # ms
        passed = False
        messages.append(f"Latency increased by: {latency_diff}ms")
        
    return passed, messages

baseline = {'accuracy': 0.95, 'latency_ms': 120}
candidate = {'accuracy': 0.945, 'latency_ms': 180}

result, msgs = evaluate_deployment(candidate, baseline)
print(f"Deployment Passed: {result}")
if msgs: print(f"Issues: {msgs}")
````
