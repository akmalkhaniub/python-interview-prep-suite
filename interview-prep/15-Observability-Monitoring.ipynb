## 1. The Three Pillars of Observability

### **Interview Question**: "What is the difference between Metrics, Logs, and Traces?"

**Answer**:
- **Metrics**: Numerical data over time (e.g., CPU % at 95). Great for **alerts** and high-level health.
- **Logs**: Discrete events (e.g., "User 123 logged in"). Great for **debugging** specific incidents.
- **Traces**: The journey of a single request through multiple services. Great for **bottleneck analysis** in microservices.

---

## 2. Structured vs. Standard Logging

### **Interview Question**: "Why should we use Structured Logging instead of simple print/string logs?"

**Answer**: Standard logs are for humans; Structured logs (usually JSON) are for **machines**. Machines can easily parse, filter, and aggregate JSON logs, allowing you to quickly query questions like: "Show me all 500 errors for User ID 789 during the last 10 minutes."

## 4. Coding Challenge: Structured Logging in Python

**Task**: Create a simple JSON logger function that mimics `structlog` by allowing developers to pass key-value context to every log message.


```python
import json
import datetime

def log_event(level, message, **kwargs):
    log_entry = {
        "timestamp": datetime.datetime.utcnow().isoformat(),
        "level": level.upper(),
        "message": message,
        **kwargs
    }
    print(json.dumps(log_entry))

# Example usage
log_event("info", "Order processed", order_id=5501, user_id="akmal_k", status="success")
log_event("error", "Database connection lost", db_host="prod-db-01", retry_count=3)
```

## 5. OpenTelemetry (OTel)

### **Interview Question**: "Why is everyone moving to OpenTelemetry?"

**Answer**: It is **vendor-neutral**. Instead of instrumenting your code with AWS X-Ray SDK or Datadog agent, you use the OTel standard. This means you can switch your monitoring backend (e.g., from CloudWatch to New Relic) without writing a single new line of application code.
