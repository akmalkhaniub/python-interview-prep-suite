## 2. Math: Stereo Projection Matrix

**Task:** Given the Field of View (FOV) angles from OpenXR, calculate a simple projection matrix.

```python
import math

def create_projection_matrix(fov_left, fov_right, fov_up, fov_down, near, far):
    # fov angles are in radians
    l = math.tan(fov_left) * near
    r = math.tan(fov_right) * near
    t = math.tan(fov_up) * near
    b = math.tan(fov_down) * near
    
    # Standard OpenGL-style perspective matrix (column-major)
    matrix = [
        [2*near/(r-l), 0, (r+l)/(r-l), 0],
        [0, 2*near/(t-b), (t+b)/(t-b), 0],
        [0, 0, -(far+near)/(far-near), -2*far*near/(far-near)],
        [0, 0, -1, 0]
    ]
    return matrix

# Example for Quest (approximate)
fov = math.radians(45)
proj = create_projection_matrix(-fov, fov, fov, -fov, 0.1, 1000.0)
print("Top-Left Matrix Element:", proj[0][0])
```
