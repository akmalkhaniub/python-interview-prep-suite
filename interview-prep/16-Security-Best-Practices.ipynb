## 1. Authentication: OAuth2 & JWT

### **Interview Question**: "What is the risk of storing sensitive user data in a JWT payload?"

**Answer**: JWT payloads are **not encrypted**; they are only Base64 encoded and signed. Anyone who captures the token can decode it and see the payload content. Only store non-sensitive identifiers (like `user_id` or `email`) in the JWT.

#### **JWT Components**:
- **Header**: Algorithm used.
- **Payload**: Claims (user data).
- **Signature**: Ensures the token hasn't been tampered with.

## 2. Secrets Management

### **Interview Question**: "How should you manage database credentials in a production cloud environment?"

**Answer**: Never use hardcoded strings or `.env` files in production. Use a **Managed Secrets Service**. The application should use an IAM role to authenticate with the secrets manager and retrieve credentials at runtime. This allows for automated **Secret Rotation** without redeploying code.

## 4. Coding Challenge: Role-Based Access Control (RBAC)

**Task**: Create a decorator that checks if the logged-in user has the required role to access a function.


```python
from functools import wraps

# Mock user object from session/token
current_user = {"username": "admin_akmal", "roles": ["admin", "editor"]}

def requires_role(required_role):
    def decorator(f):
        @wraps(f)
        def decorated_function(*args, **kwargs):
            if required_role not in current_user.get("roles", []):
                raise PermissionError(f"Missing role: {required_role}")
            return f(*args, **kwargs)
        return decorated_function
    return decorator

@requires_role("admin")
def delete_user(user_id):
    print(f"User {user_id} deleted successfully.")

try:
    delete_user(99)
except PermissionError as e:
    print(f"Access Denied: {e}")
```

## 5. OWASP Top 10 for Python

### **Interview Question**: "How do you prevent SQL Injection in a raw SQL query?"

**Answer**: Never use f-strings or string formatting to build queries. Always use **Parameterized Queries** (placeholders like `%s` or `:param`) provided by the database driver. This ensures the database treats the input as data, not as executable code.
