## 1. Game Loop with Delta Time

**Task:** Implement a loop that simulates 10 seconds of game time, updating a position with constant velocity.

```python
import time

def simulate_game(duration_sec, fps):
    pos = 0
    velocity = 5.0 # units per second
    dt = 1.0 / fps
    total_time = 0
    
    print(f"Starting simulation: FPS={fps}, Vel={velocity}")
    
    while total_time < duration_sec:
        # Update logic
        pos += velocity * dt
        total_time += dt
    
    return pos

final_pos = simulate_game(10, 60)
print(f"Final Position after 10s: {final_pos:.2f} units")
```

## 2. Finite State Machine (AI)

**Task:** Implement a simple AI that switches between IDLE, CHASE, and ATTACK based on distance to player.

```python
class EnemyAI:
    def __init__(self):
        self.state = "IDLE"
        
    def update(self, dist_to_player):
        if self.state == "IDLE":
            if dist_to_player < 10:
                self.state = "CHASE"
        elif self.state == "CHASE":
            if dist_to_player < 2:
                self.state = "ATTACK"
            elif dist_to_player > 15:
                self.state = "IDLE"
        elif self.state == "ATTACK":
            if dist_to_player >= 2:
                self.state = "CHASE"
        
        return self.state

ai = EnemyAI()
print(f"Initial: {ai.state}")
print(f"Dist 5: {ai.update(5)}")
print(f"Dist 1: {ai.update(1)}")
print(f"Dist 20: {ai.update(20)}")
```

## 3. Math: Collision Detection (Circle-Circle)

**Task:** Write a function to detect if two circles are overlapping.

```python
import math

def check_collision(c1, c2):
    # c1 = (x, y, radius)
    dx = c1[0] - c2[0]
    dy = c1[1] - c2[1]
    distance = math.sqrt(dx**2 + dy**2)
    
    return distance < (c1[2] + c2[2])

circle1 = (0, 0, 5)
circle2 = (8, 0, 5)
print(f"Collision detected: {check_collision(circle1, circle2)}")
```
