## 3. Data Transformation (FinTech Context)

**Task:** Transform a raw transaction list into a monthly aggregate for a chart.

```python
transactions = [
    {"date": "2026-01-15", "amount": 100, "category": "Food"},
    {"date": "2026-01-20", "amount": 50, "category": "Transport"},
    {"date": "2026-02-05", "amount": 200, "category": "Food"},
    {"date": "2026-02-12", "amount": 120, "category": "Shopping"}
]

def aggregate_monthly(tx_list):
    summary = {}
    for tx in tx_list:
        month = tx['date'][:7] # YYYY-MM
        summary[month] = summary.get(month, 0) + tx['amount']
    return summary

print(f"Monthly Aggregate: {aggregate_monthly(transactions)}")
```
