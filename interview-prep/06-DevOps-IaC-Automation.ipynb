## 1. Docker for Python Developers

### **Interview Question**: "How do you optimize a Python Docker image for production?"

#### **Key Talking Points**:
- **Multi-stage Builds**: Use one stage to install build-dependencies (like `gcc`, `libpq`) and another stage to copy only the final installed packages. This results in a much smaller image.
- **Base Image**: Prefer `python:3.12-slim` over `alpine` if you have complex C-dependencies (glibc issues). 
- **.dockerignore**: Ensure `.git`, `__pycache__`, and `.env` files are not included in the image.

## 2. Kubernetes: Orchestration & Triage

### **Interview Question**: "What is the difference between a Liveness Probe and a Readiness Probe?"

**Answer**:
- **Liveness Probe**: Checks if the container is alive. If it fails, K8s kills the container and starts a new one.
- **Readiness Probe**: Checks if the container is ready to handle traffic. If it fails, K8s stops sending service traffic to that pod but doesn't kill it.

**CrashLoopBackOff**: Explain that this usually means the container is crashing on start (e.g., missing environment variables, DB connection timeout).

## 3. Terraform: Infrastructure as Code (IaC)

### **Interview Question**: "How do you handle Terraform state in a team environment?"

**Answer**: Use **Remote State** (e.g., AWS S3 with DynamoDB locking). Never commit `terraform.tfstate` to Git as it contains sensitive data and causes conflicts.

## 4. Coding Challenge: GitHub Actions YAML

**Task**: Write a GitHub Actions configuration that runs `pytest` on every push to the `main` branch.


```python
workflow_example = """
name: Python CI

on:
  push:
    branches: [ main ]
  pull_request:
    branches: [ main ]

jobs:
  build:
    runs-on: ubuntu-latest
    steps:
    - uses: actions/checkout@v4
    - name: Set up Python
      uses: actions/setup-python@v5
      with:
        python-version: '3.12'
    - name: Install dependencies
      run: |
        python -m pip install --upgrade pip
        pip install -r requirements.txt
    - name: Run tests
      run: |
        pytest
"""
print("CI/CD workflow structure defined.")
```

## 5. Security & Secrets

### **Interview Question**: "How do you inject environment variables into a production Kubernetes pod?"

**Answer**: Use **Kubernetes Secrets** or better, an external vault like **HashiCorp Vault** or **AWS Secrets Manager** synced via an operator like External Secrets Operator (ESO).
