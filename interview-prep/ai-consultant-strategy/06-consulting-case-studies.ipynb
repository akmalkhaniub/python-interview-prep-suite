## 1. ROI Calculation Case Study

**Scenario:** A legal firm wants to use an LLM to summarize contracts. 
- 100 lawyers, each spends 5 hours/week summarizing.
- Avg lawyer cost: $150/hr.
- AI reduces time by 80%.
- Implementation cost: $50,000 (one-time).
- Monthly API/Token cost: $2,000.

**Task:** Calculate the Year 1 ROI.

```python
num_lawyers = 100
hours_per_week = 5
hourly_rate = 150
reduction_rate = 0.8

weekly_savings = num_lawyers * hours_per_week * hourly_rate * reduction_rate
annual_savings = weekly_savings * 52

implementation_cost = 50000
annual_token_cost = 2000 * 12
total_cost = implementation_cost + annual_token_cost

net_benefit = annual_savings - total_cost
roi_percentage = (net_benefit / total_cost) * 100

print(f"Total Year 1 Savings: ${annual_savings:,.2f}")
print(f"Total Year 1 Cost: ${total_cost:,.2f}")
print(f"Year 1 ROI: {roi_percentage:.2f}%")
```

## 2. Prioritization Matrix Exercise

**Task:** Rank the following use cases for a Retailer based on High/Medium/Low Impact and Feasibility.

1. **Inventory Demand Prediction:** High data availability, High impact on waste reduction.
2. **Store Associate Chatbot:** Low data availability, Medium impact on productivity.
3. **Personalized Email Recommendations:** Medium data availability, High impact on revenue.
4. **AI-Powered Store Security (CV):** Complex hardware requirements, Medium impact on theft reduction.

```python
import pandas as pd

data = {
    'Use Case': ['Demand Prediction', 'Associate Bot', 'Email Personalization', 'AI Security'],
    'Impact': [3, 2, 3, 2], # 3: High, 2: Med, 1: Low
    'Feasibility': [3, 1, 2, 1]
}
df = pd.DataFrame(data)
df['Score'] = df['Impact'] * df['Feasibility']
df = df.sort_values(by='Score', ascending=False)

print("Prioritized Roadmap:")
print(df)
```
