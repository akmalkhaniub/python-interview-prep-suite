## 1. Minimal RAG Implementation (Conceptual)

**Task:** Implement a simple semantic search using cosine similarity.

```python
import numpy as np

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

# Dummy embeddings (e.g., from an LLM API)
docs = [
    {"text": "The capital of France is Paris.", "embedding": [0.1, 0.9, 0.2]},
    {"text": "Apple is a technology company.", "embedding": [0.8, 0.1, 0.5]},
    {"text": "The weather is nice today.", "embedding": [0.3, 0.3, 0.9]}
]

query_embedding = [0.15, 0.85, 0.25]

def retrieve(query_emb, doc_list, top_k=1):
    scores = []
    for doc in doc_list:
        score = cosine_similarity(query_emb, doc['embedding'])
        scores.append((score, doc['text']))
    
    scores.sort(key=lambda x: x[0], reverse=True)
    return scores[:top_k]

results = retrieve(query_embedding, docs)
print(f"Top Result: {results[0][1]} (Score: {results[0][0]:.4f})")
```
