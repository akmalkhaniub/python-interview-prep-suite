## 1. Advanced Fixtures & Scoping

### **Interview Question**: "When would you use `session` scope over `function` scope for a fixture? What are the risks?"

#### **Key Talking Points**:
- **Session Scope**: The fixture is created once per test session. Ideal for expensive resources like a Database engine or an API client.
- **Function Scope**: The fixture is created for every single test. Ideal for lightweight objects where state isolation is critical.
- **The Risk**: **State Pollution**. If a session-scoped fixture is modified by one test, it can cause "flaky tests" (non-deterministic results) in subsequent tests.

## 2. Mocking & Patching

### **Interview Question**: "What is 'Over-mocking', and how do you avoid it?"

**Answer**: Over-mocking occurs when you mock internal implementation details instead of external boundaries. This results in tests that pass even when the code is broken (or break when you refactor even if logic is correct).

**The Rule**: "Mock the boundary, not the logic." Mock database calls, file systems, and network APIs. Avoid mocking internal private methods.

#### **Scenario**: Mocking an external API call using `pytest-mock`.


```python
import requests

def get_weather(city):
    response = requests.get(f"https://api.weather.com/{city}")
    return response.json()

def test_get_weather(mocker):
    # Arrange: Mock the requests.get call
    mock_response = mocker.Mock()
    mock_response.json.return_value = {"temp": 25, "condition": "Sunny"}
    mocker.patch("requests.get", return_value=mock_response)

    # Act
    result = get_weather("London")

    # Assert
    assert result["temp"] == 25
    requests.get.assert_called_once_with("https://api.weather.com/London")

print("Mocking example ready.")
```

## 3. Parametrization & Scenario Testing

### **Interview Question**: "How do you test a function against multiple edge cases without writing 10 different test functions?"

**Answer**: Use `@pytest.mark.parametrize`. It allows you to run the same test logic with different inputs and expected outputs.

#### **Example**: Testing a password validator.


```python
import pytest

def is_valid_password(pwd):
    return len(pwd) >= 8 and any(c.isdigit() for c in pwd)

@pytest.mark.parametrize("pwd,expected", [
    ("short1", False),      # Too short
    ("no_digits", False),   # No digits
    ("valid_pwd123", True), # Valid
    ("12345678", True),     # Valid digits
])
def test_password_validation(pwd, expected):
    assert is_valid_password(pwd) == expected

print("Parametrization example ready.")
```

## 4. TDD (Test-Driven Development)

### **Interview Question**: "Explain the Red-Green-Refactor cycle."

1. **RED**: Write a failing test for the next bit of functionality you want to add.
2. **GREEN**: Write just enough code to make the test pass.
3. **REFACTOR**: Clean up the code (both production and test code) while keeping the tests passing.
