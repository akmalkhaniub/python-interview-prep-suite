## 1. Bloom Filters (Probabilistic Sets)

### **Interview Question**: "How would you check if a username is already taken among 1 billion users WITHOUT scanning the database for every keystroke?"

**The Answer**: A **Bloom Filter**.
- It is a bitset that uses hashing to tell you if an item is "Definitely Not" in a set or "Maybe" in a set.
- **Pros**: Extremely fast and uses tiny memory.
- **Cons**: It can have **False Positives** (says it might be taken when it's not), but never **False Negatives**.

---

## 2. HyperLogLog (Cardinality Estimation)

### **Interview Question**: "How do you count the number of unique IP addresses visiting a site in a day (e.g., 500 million uniques) using only 1KB of memory?"

**The Answer**: **HyperLogLog**.
- It uses probabilistic counting to estimate the number of unique elements (cardinality) with an error rate of ~1%.
- It doesn't store the actual items, only a hash-based representation.

---

## 3. Merkle Trees (Hash Trees)

### **Interview Question**: "How do you verify that two massive datasets on different servers are identical without sending the whole file over the network?"

**The Answer**: **Merkle Trees**.
- You break the data into chunks, hash them, and then hash the hashes upward to a single **Root Hash**.
- If the Root Hashes match, the datasets are identical. If they don't, you can quickly find the specific chunk that differs.
- *Used in: Git, Cassandra, Bitcoin, Bittorrent.*

## 4. Coding Challenge: Simple Bloom Filter

**Task**: Implement a basic bitset-based membership check using two different hash functions.


```python
import hashlib

class SimpleBloomFilter:
    def __init__(self, size=1000):
        self.size = size
        self.bitset = [0] * size

    def _hashes(self, item):
        # Use two different hash algorithms to minimize collisions
        h1 = int(hashlib.md5(item.encode()).hexdigest(), 16) % self.size
        h2 = int(hashlib.sha1(item.encode()).hexdigest(), 16) % self.size
        return h1, h2

    def add(self, item):
        h1, h2 = self._hashes(item)
        self.bitset[h1] = 1
        self.bitset[h2] = 1

    def maybe_contains(self, item):
        h1, h2 = self._hashes(item)
        return self.bitset[h1] == 1 and self.bitset[h2] == 1

bf = SimpleBloomFilter()
bf.add("akmalkhani")

print(f"Contains 'akmalkhani': {bf.maybe_contains('akmalkhani')}")
print(f"Contains 'stranger': {bf.maybe_contains('stranger')}")
```
