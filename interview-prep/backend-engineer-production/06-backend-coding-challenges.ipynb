## 1. API: Error Handling & Validation

**Task:** Implement a function that validates a JSON payload and returns appropriate HTTP-style status codes.

```python
def create_user(payload):
    # 1. Validation
    if not payload.get("email") or "@" not in payload["email"]:
        return {"status": 400, "body": {"error": "Invalid email"}}
    
    if len(payload.get("password", "")) < 8:
        return {"status": 400, "body": {"error": "Password too short"}}
    
    # 2. Logic (Simulated)
    try:
        # In real app: db.users.insert(payload)
        return {"status": 201, "body": {"id": 101, "email": payload["email"]}}
    except Exception as e:
        return {"status": 500, "body": {"error": "Internal server error"}}

print(f"Success: {create_user({'email': 'test@example.com', 'password': 'password123'})}")
print(f"Failure: {create_user({'email': 'invalid', 'password': '123'})}")
```

## 3. Reliability: Token Bucket Rate Limiter

**Task:** Implement a simple class to track request tokens.

```python
import time

class RateLimiter:
    def __init__(self, capacity, fill_rate):
        self.capacity = capacity
        self.fill_rate = fill_rate
        self.tokens = capacity
        self.last_fill = time.time()

    def allow_request(self):
        now = time.time()
        # Refill tokens
        self.tokens += (now - self.last_fill) * self.fill_rate
        self.tokens = min(self.tokens, self.capacity)
        self.last_fill = now
        
        if self.tokens >= 1:
            self.tokens -= 1
            return True
        return False

limiter = RateLimiter(5, 1) # 5 tokens, refill 1/sec
print(f"Request 1 allowed: {limiter.allow_request()}")
```
