## 2. Performance: Python Vectorization

**Task:** Optimize a slow for-loop calculation using Numpy.

````python
import numpy as np
import time

# Scenario: Multiply two large lists element-wise
data_size = 1000000
a = list(range(data_size))
b = list(range(data_size))

# Slow way (For loop)
start = time.time()
c_slow = [a[i] * b[i] for i in range(data_size)]
print(f"For-loop time: {time.time() - start:.4f}s")

# Fast way (Numpy Vectorization)
a_np = np.array(a)
b_np = np.array(b)
start = time.time()
c_fast = a_np * b_np
print(f"Numpy time: {time.time() - start:.4f}s")
````
