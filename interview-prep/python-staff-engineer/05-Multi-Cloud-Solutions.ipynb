## 1. Serverless Architectures

### **Interview Question**: "How do you handle 'Cold Starts' in AWS Lambda or Azure Functions?"

#### **Key Talking Points**:
- **Execution Environment**: Explain that the first request to a function requires provisioning a container and starting the runtime (Cold Start).
- **Optimization**: 
    - Keep dependencies minimal (smaller image/zip size).
    - Increase allocated memory (proportionally increases CPU and networking).
    - Use **Provisioned Concurrency** (AWS) or **Premium Plan** (Azure) to keep instances warm.
    - Use specific runtimes (e.g., Python 3.12 is optimized for faster startup).

---

## 2. Cloud Storage Patterns

### **Interview Question**: "When would you use Object Storage (S3/Blob) vs. a Block Store (EBS/Disk)?"

**Answer**:
- **Object Storage**: Infinite scale, accessible via HTTP/API, perfect for unstructured data (images, logs, long-term backups). 
- **Block Storage**: Lower latency, acts like a physical hard drive, required for Databases or applications that need random read/write access to a filesystem.

---

## 3. IAM & The Principle of Least Privilege

### **Interview Question**: "How do you securely give a Python application running on EC2 access to an S3 bucket?"

**The Wrong Way**: Hardcoding Access Keys in `.env` or Git.
**The Right Way**: Use an **IAM Role** (AWS) or **Managed Identity** (Azure/GCP). The application automatically retrieves temporary credentials via the metadata service, and you define a policy that grants access *only* to the specific bucket needed.

## 4. Coding Challenge: Interacting with S3 via Boto3

**Task**: Create a Python function that uploads a local file to S3 and attaches extra metadata to the object.


```python
import boto3
from botocore.exceptions import NoCredentialsError

def upload_to_s3(file_name, bucket, s3_path):
    """
    Note: Requirement is to have AWS credentials configured locally or an IAM role.
    """
    s3 = boto3.client('s3')

    try:
        s3.upload_file(
            file_name, 
            bucket, 
            s3_path, 
            ExtraArgs={'Metadata': {'Project': 'Telemedicine', 'Type': 'Report'}}
        )
        print("Upload Successful")
    except FileNotFoundError:
        print("The file was not found")
    except NoCredentialsError:
        print("Credentials not available")

print("Cloud integration logic ready.")
```
