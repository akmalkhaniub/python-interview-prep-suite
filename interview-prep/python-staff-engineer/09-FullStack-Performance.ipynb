## 1. HTMX: Modern UI with Minimal JS

### **Interview Question**: "What is HTMX, and why would you use it instead of React for a Python project?"

**Answer**:
- **What it is**: HTMX allows you to access AJAX, WebSockets, and Server Sent Events directly in HTML attributes (e.g., `hx-get`, `hx-target`). It follows the **HATEOAS** (Hypermedia as the Engine of Application State) principle.
- **Why it's great for Python**: You keep your application logic on the server (Django/FastAPI). Instead of returning JSON and having a complex JS state machine, you return **HTML fragments** that HTMX swaps into the page.
- **Trade-off**: Great for CRUD and data-rich apps; less ideal for highly interactive client-side apps like video editors or complex drag-and-drop tools.

## 2. WebSockets & Real-time Communication

### **Interview Question**: "How do you scale WebSockets across multiple server instances?"

**The Problem**: WebSockets are stateful, long-lived TCP connections. If Client A is connected to Server 1, and Client B is connected to Server 2, they cannot talk to each other directly.

**The Solution**: Use a **Pub/Sub Broker** (usually **Redis**). When a message needs to be broadcast, Server 1 publishes it to Redis, and Server 2 (subscribed to the same channel) receives it and pushes it to Client B.

## 3. Security: JWT vs. Sessions

### **Interview Question**: "Explain the security trade-offs between using JWTs and Server-side Sessions."

| Method | Pros | Cons |
| :--- | :--- | :--- |
| **Sessions** | Secure by default (HTTPOnly cookies), easy to revoke. | Stateful (needs a DB/Redis lookup on every request). |
| **JWT** | Stateless (no DB lookup needed), works great for microservices. | Hard to revoke (blacklisting is needed), can be large (affects bandwidth). |

## 4. Coding Challenge: FastAPI WebSocket

**Task**: Create a basic WebSocket endpoint that echoes back whatever the client sends, but prepends it with "Python says:".


```python
from fastapi import FastAPI, WebSocket

app = FastAPI()

@app.websocket("/ws")
async def websocket_endpoint(websocket: WebSocket):
    await websocket.accept()
    while True:
        data = await websocket.receive_text()
        await websocket.send_text(f"Python says: {data}")

print("WebSocket endpoint logic defined.")
```

## 5. Performance Optimization

### **Interview Question**: "What are the three most important metrics for frontend performance, and how do you improve them as a full-stack dev?"

**Answer** (Core Web Vitals):
1. **LCP (Largest Contentful Paint)**: Improve by using a CDN and lazy-loading non-critical images.
2. **FID (First Input Delay)**: Improve by reducing long-running JS and using `async`/`defer` script tags.
3. **CLS (Cumulative Layout Shift)**: Improve by setting explicit height/width on images and avoiding dynamic content injections before fonts are loaded.
