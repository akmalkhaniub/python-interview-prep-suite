## 1. SOLID Principles in Python

### **Interview Question**: "Explain the 'L' in SOLID (Liskov Substitution Principle)."

**Answer**: Subtypes must be substitutable for their base types. In Python, this means a subclass should not break the contract established by the superclass (e.g., throwing a different exception or requiring different input types).

#### **The SOLID Breakdown**:
1. **S**ingle Responsibility (SRP) - A class should have one reason to change.
2. **O**pen/Closed (OCP) - Software entities should be open for extension, but closed for modification.
3. **L**iskov Substitution (LSP - as above).
4. **I**nterface Segregation (ISP) - Many client-specific interfaces are better than one general-purpose interface (implemented using Abstract Base Classes in Python).
5. **D**ependency Inversion (DIP) - Depend on abstractions, not concretions.

## 2. Creational Patterns: Singleton

### **Interview Question**: "How do you implement a thread-safe Singleton in Python?"

**Answer**: You can use a Metaclass or the `__new__` method. Using a Metaclass is the most robust way.

#### **Example**: Singleton using `__new__`.


```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(Singleton, cls).__new__(cls)
        return cls._instance

s1 = Singleton()
s2 = Singleton()
print(f"Are s1 and s2 the same? {s1 is s2}")
```

## 3. Behavioral Patterns: Strategy

### **Interview Question**: "What is the Strategy pattern, and how does it help with the 'Open/Closed' principle?"

**Answer**: The Strategy pattern allows you to define a family of algorithms, encapsulate each one, and make them interchangeable. It allows the algorithm to vary independently from clients that use it. It helps OCP because you can add new strategies without modifying existing code.

#### **Example**: Payment Processing Strategy.


```python
from abc import ABC, abstractmethod

class PaymentStrategy(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paying {amount} using Credit Card")

class PayPalPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Paying {amount} using PayPal")

class ShoppingCart:
    def __init__(self, strategy: PaymentStrategy):
        self.strategy = strategy

    def checkout(self, amount):
        self.strategy.pay(amount)

cart = ShoppingCart(PayPalPayment())
cart.checkout(450)

print("Strategy pattern demonstration ready.")
```

## 4. The Repository Pattern

### **Interview Question**: "Why is a Repository pattern useful when working with an ORM like SQLAlchemy?"

**Answer**: It decouples the business logic from the data access layer. It makes it easier to swap the database for testing (e.g., using an in-memory list) and keeps the domain layer clean of ORM-specific code.
