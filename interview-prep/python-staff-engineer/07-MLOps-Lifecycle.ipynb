## 1. Feature Stores & The "Skew"

### **Interview Question**: "What is Training-Serving Skew, and how does a Feature Store prevent it?"

**Answer**:
- **Training-Serving Skew**: When the features used for training differ from the features available during real-time inference (e.g., different preprocessing logic or data snapshots).
- **Feature Store**: Acts as a central repository (e.g., **Feast**, **Tecton**). It ensures that the exact same feature engineering code is used for both offline training and online serving.

---

## 2. Model Tracking & Registry

### **Interview Question**: "Why do we need a Model Registry? Can't we just save our `.pkl` files in S3?"

**The Problem with S3-only**: No versioning, no metadata association (which dataset was used?), no lifecycle management (Staging -> Production).

**The Solution**: A **Model Registry** (e.g., **MLflow**, **Weights & Biases**). It tracks hyperparameters, model metrics, and allows for "one-click" rollbacks to previous versions.

---

## 3. Data Drift vs. Concept Drift

### **Interview Question**: "How do you detect when your model is becoming less accurate over time?"

**Definitions**:
- **Data Drift**: The statistical distribution of the *input* data changes (e.g., a new age group starts using your app).
- **Concept Drift**: The relationship between the input and the *target* changes (e.g., consumer behavior shifts during a pandemic).

**Detection Tools**: Mention **Evidently AI**, **WhyLogs**, or custom Kolmogorov-Smirnov (K-S) tests on incoming feature distributions.

## 4. Coding Challenge: Mock MLflow Tracking

**Task**: Simulate a model training run that logs parameters and metrics.


```python
import random

class MockMLflow:
    def __init__(self):
        self.logs = []

    def log_param(self, name, value):
        self.logs.append(f"PARAM: {name} = {value}")

    def log_metric(self, name, value):
        self.logs.append(f"METRIC: {name} = {value}")

    def show_run(self):
        print("\n--- MLflow Run Summary ---")
        for log in self.logs:
            print(log)

mlflow = MockMLflow()

# Simulate training
params = {"learning_rate": 0.01, "epochs": 10, "batch_size": 32}
for p, v in params.items():
    mlflow.log_param(p, v)

accuracy = 0.85 + (random.random() * 0.1)
mlflow.log_metric("accuracy", round(accuracy, 4))

mlflow.show_run()
```

## 5. Deployment Patterns

### **Interview Question**: "How do you perform a Shadow Deployment of a new model?"

**Answer**: "In a **Shadow Deployment**, the new model receives the same traffic as the production model but its predictions are *not* returned to the user. We log both model outputs to compare accuracy/latency in a production environment without risk."
