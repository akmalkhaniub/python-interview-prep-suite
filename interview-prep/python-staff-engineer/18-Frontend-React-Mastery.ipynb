## 1. The Virtual DOM & Reconciliation

### **Interview Question**: "How does React efficiently update the UI without re-rendering the whole page?"

**Answer**: React uses a **Virtual DOM**. 
1. When data changes, React creates a new Virtual DOM tree.
2. It compares the new tree with the old tree (**Diffing**).
3. It identifies the exact changes needed and applies them to the real DOM (**Reconciliation**).
This is much faster than manual DOM manipulation.

## 2. Rendering Strategies: SSR vs SSG vs CSR

### **Interview Question**: "When would you use Server-Side Rendering (SSR) over Client-Side Rendering (CSR)?"

**Answer**:
- **CSR**: Standard React apps. Fast navigation after first load, but poor SEO and slow initial load.
- **SSR**: Best for SEO and fast "Time to First Byte". The server renders the HTML and sends it to the browser.
- **SSG (Static Site Generation)**: Best for content that doesn't change often (blogs, docs). Pre-rendered at build time.

---

## 3. State Management Trade-offs

### **Interview Question**: "Redux vs Context API vs Zustand: Which one should I choose?"

**Guidelines**:
- **Context API**: Best for small/medium apps and static data (themes, auth status). Avoid for highly dynamic data due to re-render performance issues.
- **Redux**: Best for massive, complex apps with strictly defined data flows. Overkill for many modern projects.
- **Zustand**: The modern favorite. Lightweight, no boilerplate, and very fast. Great balance for most apps.

## 4. Performance Optimization

### **Interview Question**: "How do you prevent unnecessary re-renders in a React component?"

**The Tools**:
- **React.memo**: Memoizes the whole component.
- **useMemo**: Memoizes the *result* of a calculation.
- **useCallback**: Memoizes the *function instance* itself (useful when passing callbacks to children).
