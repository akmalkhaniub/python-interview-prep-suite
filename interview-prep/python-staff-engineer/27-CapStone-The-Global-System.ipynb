## 2. The Data Path: Finding "Neighbors"

### **The Complexity**: How do you index millions of moving objects?

**The Solution**: **Geohashes** or **Quadtrees**.
1. **Geohash**: A base32 string representing a bounding box on earth. A longer string means a smaller area.
2. **Algorithm**: Store provider locations in **Redis** indexed by their Geohash prefix. 
3. **Lookup**: When a patient requests a match, we calculate their Geohash and query Redis for providers with the same prefix.

## 5. Coding Challenge: The Geohash Logic

**Task**: Implement a simple function that simulates finding providers within a specific Geohash "bucket".


```python
providers = [
    {"id": 1, "geohash": "dr5rs1"},
    {"id": 2, "geohash": "dr5rs2"},
    {"id": 3, "geohash": "9q9h12"}, # In a different city
]

def find_nearby_providers(user_geohash, provider_list):
    # We match on the first 5 characters for 'city-wide' search
    prefix = user_geohash[:5]
    results = [p for p in provider_list if p["geohash"].startswith(prefix)]
    return results

nearby = find_nearby_providers("dr5rsx", providers)
print(f"Found {len(nearby)} providers in your city.")
```
