## 1. Distributed Training Setup (DDP)

**Task:** Outline the boilerplate code to initialize a Distributed Data Parallel (DDP) environment in PyTorch.

```python
import torch
import torch.distributed as dist
from torch.nn.parallel import DistributedDataParallel as DDP

def setup(rank, world_size):
    # 1. Init process group
    dist.init_process_group("nccl", rank=rank, world_size=world_size)
    
    # 2. Set device
    torch.cuda.set_device(rank)
    
    # 3. Model to DDP
    model = torch.nn.Linear(10, 10).to(rank)
    ddp_model = DDP(model, device_ids=[rank])
    
    return ddp_model

print("DDP Setup logic defined.")
```

## 4. Optimization: Memory Savings (Quantization)

**Task:** Calculate the VRAM savings when moving a 70B model from FP16 to 4-bit.

```python
params = 70e9
fp16_bytes = 2
q4_bytes = 0.5 # 4 bits = 0.5 bytes

fp16_vram = (params * fp16_bytes) / 1e9 # in GB
q4_vram = (params * q4_bytes) / 1e9

print(f"70B Model VRAM (FP16): {fp16_vram:.2f} GB")
print(f"70B Model VRAM (4-bit): {q4_vram:.2f} GB")
print(f"Savings: {fp16_vram - q4_vram:.2f} GB")
```
