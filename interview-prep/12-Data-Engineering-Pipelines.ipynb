## 1. ETL vs. ELT Philosophy

### **Interview Question**: "When would you choose ELT over traditional ETL?"

**Answer**:
- **ETL (Extract, Transform, Load)**: Data is cleaned *before* entering the warehouse. Best for PII masking, strict compliance, or when you want to save on warehouse storage costs by filtering data early.
- **ELT (Extract, Load, Transform)**: Data is loaded raw into the warehouse and transformed using the warehouse's compute (e.g., Snowflake, BigQuery). Best for agility, large-scale data, and when you want to keep raw data for future re-processing.

## 2. PySpark Performance Tuning

### **Interview Question**: "What is Data Skew, and how do you fix it in Spark?"

**Definition**: When most of your data is concentrated in a few partitions, causing them to take much longer to process (The "Straggler" problem).

**Solutions**:
1. **Salting**: Adding a random prefix to the join keys to distribute the data more evenly across partitions.
2. **Broadcast Joins**: If one table is small, broadcast it to all executors to avoid a shuffle.
3. **Adaptive Query Execution (AQE)**: Enabling Spark's built-in feature to handle skew dynamically.

## 4. Coding Challenge: Idempotent Processing

**Task**: Write a PySpark snippet that reads data and ensures no duplicate records are processed (idempotency).


```python
# Note: This requires a SparkSession (spark)
from pyspark.sql import functions as F
from pyspark.sql.window import Window

def clean_duplicates(df, key_column):
    """
    Keeps the most recent record for each key based on a timestamp.
    """
    window_spec = Window.partitionBy(key_column).orderBy(F.col("timestamp").desc())
    
    return df.withColumn("row_num", F.row_number().over(window_spec)) \
             .filter(F.col("row_num") == 1) \
             .drop("row_num")

print("Idempotent processing logic ready.")
```

## 5. Metadata-Driven Pipelines

### **Interview Question**: "What is a metadata-driven pipeline, and why is it superior to hard-coded ETL?"

**Answer**: A metadata-driven pipeline uses a configuration file (JSON/YAML) to define the source and destination schema. This allows you to add new tables without writing new code, and makes it easier to handle schema evolution and automated data quality checks.
