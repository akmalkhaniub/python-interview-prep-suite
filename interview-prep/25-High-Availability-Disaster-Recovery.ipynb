# Interview Prep: 25 - High Availability & Disaster Recovery

Scaling is about success; High Availability (HA) is about **surviving failure**. This notebook covers how to design systems that keep running even when an entire data center goes dark.

---

## 1. HA vs. DR: The Big Difference

### **Interview Question**: "What is the difference between High Availability and Disaster Recovery?"

**The Answer**:
- **High Availability (HA)**: Redundancy *within* a system to prevent downtime during minor failures (e.g., a single server crashing). Usually involves multiple **Availability Zones (AZs)**.
- **Disaster Recovery (DR)**: The plan for surviving a catastrophic event (e.g., an entire AWS Region going offline). Usually involves multiple **Regions**.

---

## 2. RTO and RPO

### **Interview Question**: "A business stakeholder says we need 'zero data loss' and 'zero downtime'. How do you respond?"

**The Answer**: Discuss the **Cost vs. Reliability** trade-off using these metrics:
- **RPO (Recovery Point Objective)**: How much data loss can we tolerate? (e.g., 5 minutes of logs).
- **RTO (Recovery Time Objective)**: How long can we be offline? (e.g., 2 hours).
"Zero/Zero" is theoretically possible (Active-Active Multi-region) but is **extremely expensive** to build and maintain.

---

## 3. Multi-Region Patterns

### **Interview Question**: "When would you choose Active-Active over Active-Passive for a global database?"

**The Answer**:
- **Active-Passive**: Traffic goes to Region A. Region B is a sync-standby. Simple but has a downtime gap during failover.
- **Active-Active**: Traffic goes to both regions. Users are routed to the closest one. Requires complex **Conflict Resolution** but provides near-zero RTO.

---

## 4. Chaos Engineering

### **Interview Question**: "How do you trust that your failover logic actually works?"

**The Answer**: **Chaos Engineering**.
- You must intentionally break things in production (or staging) to verify the system recovers.
- **Chaos Monkey**: A tool that randomly kills production instances to ensure the auto-scaling group and load balancer handle the loss gracefully.

## 5. Coding Scenario: Health Check Logic

**Task**: Write a heartbeat function that handles a "Brownout" (partial failure) rather than just a total crash.


```python
def check_system_health(api_latency, error_rate):
    """
    Complex health check for a load balancer.
    """
    if error_rate > 0.1: # 10% errors
        return "CRITICAL: Immediate Failover"
    if api_latency > 2.0: # 2s latency
        return "DEGRADED: Stop sending new traffic"
    return "HEALTHY"

print(f"Status (Latency 2.5s): {check_system_health(2.5, 0.02)}")
```
