## 3. Developer Feedback Prioritization

**Task:** You have 4 pieces of feedback. Which one do you take to the Engineering team first?

1. **"Documentation says function X takes a string, but it actually needs an object."** (High Friction)
2. **"I want a dark mode for the SDK website."** (Low Impact)
3. **"The SDK crashes on Windows with AMD GPUs."** (High Criticality)
4. **"Can we add a 'summarize' method to the core API?"** (Feature Request)

````python
priority = [
    (1, "AMD GPU Crash", "Blocks entire segment of users (Criticality)."),
    (2, "Doc Typo", "Fast fix, prevents onboarding friction (Quick Win)."),
    (3, "Summarize Method", "Valuable, but should be planned in the roadmap (Product Strategy)."),
    (4, "Dark Mode", "Purely aesthetic, lowest priority.")
]

print("Priority List:")
for rank, item, reason in priority:
    print(f"{rank}. {item}: {reason}")
````
