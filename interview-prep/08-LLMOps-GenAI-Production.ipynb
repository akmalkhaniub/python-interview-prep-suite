## 1. RAG Architecture & Optimization

### **Interview Question**: "Explain your strategy for chunking and indexing high-volume documents for a RAG pipeline."

#### **Key Talking Points**:
- **Chunking**: Decision between fixed-size (e.g., 512 tokens) vs. semantic (based on headers/paragraphs). Mention **recursive character splitting**.
- **Indexing**: Embedding models (e.g., `text-embedding-3-small`) and Vector DB selection (e.g., Pinecone for managed, Chroma for local, pgvector for relational integrity).
- **Hybrid Search**: Combining Vector search (semantic) with Keyword search (BM25) to handle specific jargon or product names.

## 2. Evaluation: The RAG Triad

### **Interview Question**: "How do you move beyond 'vibe checks' and systematically evaluate your LLM application?"

**The RAG Triad** (popularized by TruLens/Ragas):
1. **Context Relevance**: Is the retrieved context actually useful for answering the query?
2. **Faithfulness (Groundedness)**: Is the answer derived *only* from the retrieved context (no hallucinations)?
3. **Answer Relevance**: Does the final answer directly address the user's prompt?

#### **Tooling**: Mention **Ragas**, **DeepEval**, or using an "LLM-as-a-judge" (e.g., GPT-4o evaluating a Llama 3 response).

## 4. Coding Challenge: Answer Relevancy Metric

**Task**: Implement a simple function that calculates a basic overlap score between a query and an answer as a proxy for relevancy.


```python
import re
from collections import Counter

def calculate_answer_relevancy(query: str, answer: str):
    """
    A simple word-overlap baseline for relevancy checking.
    In production, use cosine similarity of embeddings or LLM-as-a-judge.
    """
    def get_tokens(text):
        return set(re.findall(r'\w+', text.lower()))

    query_tokens = get_tokens(query)
    answer_tokens = get_tokens(answer)

    intersection = query_tokens.intersection(answer_tokens)
    score = len(intersection) / len(query_tokens) if query_tokens else 0
    
    return score

query = "How does async work in Python?"
answer = "Async in Python uses an event loop to handle concurrent I/O tasks."

print(f"Relevancy Score: {calculate_answer_relevancy(query, answer):.2f}")
```

## 5. Security: Prompt Injection

### **Interview Question**: "How do you prevent a user from 'jailbreaking' your LLM via a prompt?"

**Strategies**:
- **System Prompt Hardening**: Explicit instructions to ignore user overrides.
- **Output Filtering**: Using a separate LLM or regex to flag inappropriate responses.
- **Prompt Shields**: Dedicated tools (e.g., Azure Prompt Shields) that detect malicious instructions before they reach the main model.
