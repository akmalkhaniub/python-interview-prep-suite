## 1. PostgreSQL: Performance & Tuning

### **Interview Question**: "Your query is slow in production. Walk me through your triage process."

#### **Key Steps**:
1. **Identify the Query**: Use `pg_stat_statements` to find the most time-consuming queries.
2. **EXPLAIN ANALYZE**: Run the query with `EXPLAIN (ANALYZE, BUFFERS)`. Look for `Seq Scan` (Table scan) vs `Index Scan`.
3. **Check Statistics**: Are the table statistics stale? Run `ANALYZE`.
4. **Indexing**: Would a multi-column index, partial index, or a different index type (GIN/GiST) help?

**Core Concept**: **MVCC** (Multi-Version Concurrency Control). Explain that updates in Postgres don't change data in place; they create a new version, requiring **VACUUM** to clean up "dead tuples."

## 2. MongoDB: Modeling for Scale

### **Interview Question**: "When do you embed documents vs. using references?"

#### **Decision Matrix**:
- **Embed**: If the data is frequent-access, small, and related (e.g., product details in an order). Use this to avoid joins and leverage atomicity.
- **Reference**: If the data is large (>16MB per document), frequently updated independently, or used across many different entities (e.g., User profiles in many blog posts).

**Sharding**: Explain that sharding allows horizontal scaling. A **Shard Key** must have high cardinality to avoid "hot shards."

## 3. Vector Databases: The AI Engine

### **Interview Question**: "What is the difference between HNSW and IVF algorithms in Vector Search?"

**HNSW (Hierarchical Navigable Small World)**:
- **Pros**: Extremely fast retrieval, high accuracy.
- **Cons**: Memory-intensive (uses a graph structure in RAM).

**IVF (Inverted File Index)**:
- **Pros**: Low memory footprint, fast clustering.
- **Cons**: Requires periodic training (rebuilding clusters) and can be slower for very high-dimensional data.

**Hybrid Search**: Combining Dense Vectors (semantic) with Sparse Vectors (keyword search like BM25) for better accuracy on specific terms.

## 4. Coding Challenge: Simulated Index Efficiency

**Task**: Below is a Python simulation of how a Sequential Scan vs. Hash-based Lookup (Index) performs as the dataset grows.


```python
import time
import random

data_size = 1_000_000
target_id = 999_999

# 1. Sequential Scan Simulation (List lookup)
my_list = list(range(data_size))
start = time.time()
for item in my_list:
    if item == target_id:
        break
seq_time = time.time() - start

# 2. Indexed Lookup Simulation (Dictionary/Hash lookup)
my_index = {i: True for i in range(data_size)}
start = time.time()
_ = my_index.get(target_id)
idx_time = time.time() - start

print(f"Sequential Scan: {seq_time:.6f}s")
print(f"Indexed Lookup: {idx_time:.6f}s")
print(f"Speedup: {seq_time / idx_time:.1f}x")
```
