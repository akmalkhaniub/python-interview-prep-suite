## 2. Python: AI API Integration (Structured Output)

**Task:** Implement a function that calls an LLM and parses the JSON response safely.

````python
import json

def process_ai_response(raw_text):
    try:
        # In real app: response = client.completions.create(...)
        # Assume raw_text is the string returned from the API
        data = json.loads(raw_text)
        
        # Validation
        required_keys = ["sentiment", "summary", "suggested_action"]
        if all(key in data for key in required_keys):
            return {"status": "success", "data": data}
        else:
            return {"status": "error", "message": "Missing keys"}
            
    except json.JSONDecodeError:
        return {"status": "error", "message": "Invalid JSON"}
    except Exception as e:
        return {"status": "error", "message": str(e)}

mock_response = '{"sentiment": "positive", "summary": "User likes the product", "suggested_action": "send_thank_you"}'
print(process_ai_response(mock_response))
````

## 3. Data Pipeline: Simple ETL

**Task:** Extract data from a list of dicts, transform (lowercase email), and load into a "mock" DB.

````python
raw_users = [
    {"id": 1, "email": "ALICE@example.com", "name": "Alice"},
    {"id": 2, "email": "BOB@EXAMPLE.COM", "name": "Bob"}
]

def etl_pipeline(data):
    processed = []
    for entry in data:
        # Transform
        entry['email'] = entry['email'].lower()
        processed.append(entry)
    
    # Load (Mock)
    print(f"Loading {len(processed)} users into DB...")
    return processed

clean_data = etl_pipeline(raw_users)
print(clean_data)
````
