# Interview Prep: 19 - Mobile Architectures (RN vs. Flutter vs. Native)

Choosing a mobile strategy is a critical business decision. This notebook compares **React Native**, **Flutter**, and **Native (Swift/Kotlin)** development to help you navigate technical design discussions.

---

## 1. React Native: The New Architecture

### **Interview Question**: "What is JSI and how does it improve React Native performance?"

**Answer**: In the old architecture, JS and Native communicated via an asynchronous **Bridge** (JSON-based). This was slow and caused frame drops. 
- **JSI (JavaScript Interface)**: Replaces the Bridge. It allows JavaScript to hold a direct reference to C++ objects. Communication is now **synchronous** and much faster.
- **Fabric**: The new rendering engine that leverages JSI for high-priority UI updates.

---

## 2. Flutter: The Graphics Engine

### **Interview Question**: "How does Flutter's rendering differ from React Native?"

**Answer**:
- **React Native**: Maps JS components to **Native UI components** (e.g., a native iOS UIButton).
- **Flutter**: Does not use native components. It uses its own graphics engine (**Skia** or the newer **Impeller**) to draw every pixel on the screen manually. 
**Result**: Flutter has incredibly consistent animations across devices but can feel "non-native" if default styles aren't tuned carefully.

## 3. Comparison Matrix

| Feature | React Native | Flutter | Native (Swift/Kotlin) |
| :--- | :--- | :--- | :--- |
| **Language** | JavaScript/TypeScript | Dart | Swift / Kotlin |
| **Performance** | Near-native (JSI) | High (Skia/Impeller) | Peak Performance |
| **UI Components**| Native Platform | Proprietary/Custom | Native Platform |
| **Code Sharing** | High (~90%) | Very High (~95%) | Low (Platform Specific) |
| **OTA Updates** | Supported (Expo/CodePush)| Not Supported | Not Supported |

## 4. When to Choose Native?

### **Interview Question**: "Why would a company ever choose to build two separate native apps instead of using React Native?"

**Answer**: 
1. **Heavy Resource Usage**: Video editing, AR/VR, or heavy computation where GC (Garbage Collection) pauses in JS/Dart would kill the UX.
2. **Cutting-edge Features**: Native always gets new iOS/Android APIs first.
3. **OS-specific Integrations**: Deep integration with platform features (Widgets, Home Screen, Bluetooth peripherals) is often easier in native.

---

## 5. Coding Scenario: Sharing Business Logic

**Task**: Explain how you would structure a project to share 100% of your validation logic and API client between a Web app (React) and a Mobile app (React Native).

**The Solution**: Create a **Shared Library** (Monorepo or NPM package) containing pure TypeScript logic. Both the React and React Native projects import this library. Since the logic doesn't touch the DOM or Native UI, it works perfectly across both.
