# Typed Python
Python is a dynamically typed language, but since version 3.5, it has supported optional **Type Hints** (PEP 484). Type hints make code more readable, enable better IDE support (autocompletion), and allow static analysis tools like **Mypy** to find bugs before your code even runs.

## Basic Type Hints
You can annotate variables and function signatures using the `name: type` syntax.

```python
def greet(name: str, age: int) -> str:
    return f"Hello {name}, you are {age} years old."

message: str = greet("Alex", 25)
print(message)
```

## Generic Collections (PEP 585)
In modern Python (3.9+), you can use built-in types like `list`, `dict`, and `set` as generics.

```python
from typing import Any

# A list of integers
numbers: list[int] = [1, 2, 3]

# A dictionary with string keys and integer values
prices: dict[str, float] = {"apple": 1.5, "banana": 0.99}

# Using Any when the type is unknown or mixed
mixed: list[Any] = [1, "apple", True]
```

## Union and Optional
Sometimes a variable can be one of several types. Since Python 3.10, you can use the pipe operator `|` for Unions.

```python
def process_id(user_id: int | str) -> None:
    print(f"Processing ID: {user_id}")

process_id(101)
process_id("user_v1")

# Optional types (something that can be a type or None)
from typing import Optional

def get_name(id: int) -> str | None:
    if id == 1:
        return "Alex"
    return None
```

## Protocols (Static Duck Typing)
A **Protocol** defines an interface. Any class that implements the required methods is considered to "implement" the Protocol, even without explicit inheritance.

```python
from typing import Protocol

class Drawable(Protocol):
    def draw(self) -> None:
        ...

class Circle:
    def draw(self) -> None:
        print("Drawing a circle")

class Square:
    def draw(self) -> None:
        print("Drawing a square")

def render(shape: Drawable) -> None:
    shape.draw()

render(Circle())
render(Square())
```

## Generics (TypeVar)
Use `TypeVar` to create functions or classes that work with multiple types while maintaining type safety.

```python
from typing import TypeVar, Sequence

T = TypeVar("T")

def get_first(items: Sequence[T]) -> T:
    return items[0]

n: int = get_first([1, 2, 3])
s: str = get_first(["a", "b", "c"])

print(n, s)
```

## Summary
- **Variables**: `x: int = 5`
- **Functions**: `def f(a: int) -> str:`
- **Unions**: `int | str` (Python 3.10+)
- **Generics**: `list[int]`, `dict[str, Any]`
- **Protocols**: Define interfaces without inheritance.
- **Production Tip**: Always run `mypy .` in your project to verify that your type hints are correct!
