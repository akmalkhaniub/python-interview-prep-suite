# Structural Pattern Matching
Introduced in Python 3.10 (PEP 634), Structural Pattern Matching uses the `match` and `case` statements. It is much more powerful than a simple switch/case statement found in other languages, as it can deconstruct complex data structures like lists, dictionaries, and classes.

## Basic Match
The simplest use case is matching against literal values.

```python
def http_status(status):
    match status:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500 | 501 | 502:
            return "Server Error"
        case _:
            return "Unknown Status"

print(http_status(200))
print(http_status(501))
print(http_status(999))
```

## Matching Sequences
You can match against the structure and content of lists or tuples.

```python
def process_command(command):
    match command.split():
        case ["quit"]:
            print("Quitting...")
        case ["load", filename]:
            print(f"Loading {filename}")
        case ["move", x, y]:
            print(f"Moving to ({x}, {y})")
        case ["save", *files]:
            print(f"Saving {len(files)} files: {files}")
        case _:
            print("Unknown command")

process_command("load data.csv")
process_command("save file1.txt file2.txt file3.txt")
```

## Matching Dictionaries
When matching dictionaries, the pattern only needs to match a subset of the keys.

```python
def handle_event(event):
    match event:
        case {"type": "click", "button": "left"}:
            print("Left click detected")
        case {"type": "keypress", "key": key}:
            print(f"Key pressed: {key}")
        case _:
            print("System event")

handle_event({"type": "click", "button": "left", "timestamp": 12345})
handle_event({"type": "keypress", "key": "Enter"})
```

## Matching Class Instances
You can match against class attributes. This is extremely useful for processing objects in complex business logic.

```python
class Point:
    __match_args__ = ("x", "y") # Allows positional matching
    def __init__(self, x, y):
        self.x = x
        self.y = y

def location(point):
    match point:
        case Point(x=0, y=0):
            print("At the origin")
        case Point(x=0, y=y):
            print(f"On the Y-axis at {y}")
        case Point(x, 0): # Positional matching thanks to __match_args__
            print(f"On the X-axis at {x}")
        case Point(x, y):
            print(f"Somewhere at ({x}, {y})")

location(Point(0, 0))
location(Point(5, 0))
location(Point(3, 4))
```

## Guards
You can add an `if` clause to a case to perform more complex checks.

```python
def check_age(data):
    match data:
        case {"name": name, "age": age} if age < 18:
            print(f"{name} is a minor")
        case {"name": name, "age": age}:
            print(f"{name} is an adult")

check_age({"name": "Bob", "age": 15})
check_age({"name": "Alice", "age": 30})
```

## Summary
- `match subject:`: Starts a matching block.
- `case pattern:`: Defines a specific structure to match.
- `_`: Wildcard for "match anything".
- `|`: OR operator for multiple values in one case.
- **Pro-tip**: Use Pattern Matching to simplify deep nested `if-elif-else` blocks when dealing with complex data structures like API responses.
