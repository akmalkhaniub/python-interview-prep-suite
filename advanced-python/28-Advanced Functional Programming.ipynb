# Advanced Functional Programming
Python is a multi-paradigm language that supports many functional programming patterns. Libraries like `functools` and `itertools` provide powerful tools to write concise, declarative, and performant code.

## Partial Functions
`functools.partial` allows you to "freeze" some portion of a function's arguments and/or keywords, creating a new function with a simplified signature.

```python
from functools import partial

def power(base, exponent):
    return base ** exponent

square = partial(power, exponent=2)
cube = partial(power, exponent=3)

print(square(5)) # 25
print(cube(5))   # 125
```

## Single Dispatch
`functools.singledispatch` provides a way to implement **Ad-hoc Polymorphism**. It allows a single function to have different implementations based on the type of its first argument.

```python
from functools import singledispatch

@singledispatch
def process(payload):
    raise NotImplementedError("Unsupported type")

@process.register
def _(payload: int):
    print(f"Processing integer: {payload}")

@process.register
def _(payload: list):
    print(f"Processing list of length {len(payload)}")

process(10)
process([1, 2, 3])
```

## LRU Cache (Memoization)
`functools.lru_cache` is a decorator that wraps a function with a memoizing callable which saves the results of function calls and returns the cached result when the same inputs occur again.

```python
from functools import lru_cache
import time

@lru_cache(maxsize=128)
def expensive_computation(n):
    time.sleep(2) # Simulate work
    return n * n

start = time.perf_counter()
print(expensive_computation(4))
print(f"First call time: {time.perf_counter() - start:.2f}s")

start = time.perf_counter()
print(expensive_computation(4))
print(f"Second call time: {time.perf_counter() - start:.2f}s")
```

## Helpful Itertools
- `itertools.chain()`: Flattens multiple iterables.
- `itertools.compress()`: Filters an iterable based on a selector iterable.

```python
from itertools import chain, compress

list1 = [1, 2]
list2 = [3, 4]
combined = list(chain(list1, list2))
print(combined) # [1, 2, 3, 4]

data = ['A', 'B', 'C', 'D']
selectors = [True, False, True, False]
result = list(compress(data, selectors))
print(result) # ['A', 'C']
```

## Summary
- **Partial**: Prefill arguments for functions.
- **Single Dispatch**: Type-based function overloading.
- **LRU Cache**: Easy performance boost for idempotent functions.
- **Itertools**: The "standard library's high-performance loop tools".
